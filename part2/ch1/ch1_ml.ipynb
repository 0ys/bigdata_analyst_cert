# 첫번째 머신러닝 모델 만들기

```python
# 사이킷런에서 제공하는 데이터셋
import pandas as pd
from sklearn import datasets

iris = datasets.load_iris()
df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
target = iris['target']
```

```python
# 1. 문제 정의
# 다중분류, target(3개), 평가방식: 정확도(accuracy)

# 2. 라이브러리 및 데이터 불러오기
# 사이킷런에서 제공하는 데이터 활용

# 3. 탐색적 데이터 분석
print("데이터 크기:", df.shape)
print(df.head(3))

# 4. 데이터 전처리(생략)

# 5. 검증 데이터 나누기
from sklearn.model_selection import train_test_split # 데이터 분할
X_train, X_val, y_train, y_val = train_test_split(df, target, test_size=0.2)

# 6. 모델 학습 및 평가
from sklearn.ensemble import RandomForestClassifier # 머신러닝 모델(랜덤포레스트)
model = RandomForestClassifier() # 랜덤포레스트 모델 생성
model.fit(X_train, y_train) # 모델 학습
y_pred = model.predict(X_val) # 예측

from sklearn.metrics import accuracy_score # 평가(정확도)
accuracy = accuracy_score(y_val, y_pred) # 정확도 평가
print('정확도:', accuracy)

# 7. test데이터 예측 및 제출(생략)
```
