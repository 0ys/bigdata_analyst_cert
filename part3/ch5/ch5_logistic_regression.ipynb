# 로지스틱 회귀분석

```python
import pandas as pd
# df = pd.read_csv("health_survey.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part3/ch5/health_survey.csv")

print(df.head())
```

```python
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df['disease'], df['smoker'])
print(chi2_contingency(contingency_table))
```

```python
from statsmodels.formula.api import logit

model = logit('disease ~ age + bmi', data=df).fit()
print(model.summary())
```

```python
import numpy as np
print(model.params['bmi'])
print(np.exp(model.params['bmi']))
```

```python
print(model.llf)
```

```python
print(-2 * model.llf)
```
