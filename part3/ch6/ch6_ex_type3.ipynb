```python
# 데이터
import pandas as pd
df = pd.DataFrame({
    'Caffeine(mg)': [
        94.2, 93.7, 95.5, 93.9, 94.0, 95.2, 94.7, 93.5, 92.8, 94.4,
        93.8, 94.6, 93.3, 95.1, 94.3, 94.9, 93.9, 94.8, 95.0, 94.2,
        93.7, 94.4, 95.1, 94.0, 93.6
    ]
})
```

```python
# 1. 표본 평균
print(df.mean())

# 2. 정규성 검정
from scipy import stats
print(stats.shapiro(df['Caffeine(mg)']))

# 3~5.단일 표본 t-검정
print(stats.ttest_1samp(df['Caffeine(mg)'], 95, alternative='less'))
```

```python
statistic, pvalue = stats.ttest_1samp(df['Caffeine(mg)'], 95, alternative='less')
print("{:.10f}".format(pvalue))
```

```python
# 데이터
import pandas as pd
df = pd.DataFrame({
    '충전기': ['New'] * 10 + ['Old'] * 10,
    '충전시간': [
        1.5, 1.6, 1.4, 1.7, 1.5, 1.6, 1.7, 1.4, 1.6, 1.5,
        1.7, 1.8, 1.7, 1.9, 1.8, 1.7, 1.8, 1.9, 1.7, 1.6
    ]
})
print(df.head(2))
```

```python
# 1~3. 독립 표본 t-검정
new_cond = df['충전기'] == 'New'
old_cond = df['충전기'] == 'Old'
print(stats.ttest_ind(df[new_cond]['충전시간'], df[old_cond]['충전시간'], alternative='less', equal_var=True))
```

```python
new_cond = df['충전기'] == 'New'
old_cond = df['충전기'] == 'Old'

# 정규성 검정
print(stats.shapiro(df[new_cond]['충전시간']))
print(stats.shapiro(df[old_cond]['충전시간']))

# 등분산 검정
print(stats.levene(df[new_cond]['충전시간'], df[old_cond]['충전시간']))

# 독립 표본 t-검정
print(stats.ttest_ind(df[new_cond]['충전시간'], df[old_cond]['충전시간'], alternative='less', equal_var=True))
```

### Section4.

```python
# 데이터
import pandas as pd
df = pd.DataFrame({
    'group_A': [85, 88, 90, 82, 87, 89, 92, 86, 88, 90],
    'group_B': [78, 80, 79, 81, 80, 78, 77, 79, 78, 82],
    'group_C': [80, 82, 85, 84, 81, 86, 82, 81, 83, 80],
    'group_D': [85, 84, 86, 87, 85, 86, 84, 85, 87, 86]
})
print(df.head())
```

```python
from scipy import stats

# 1. Shapiro-Wilk 검정 (정규성)
print(stats.shapiro(df['group_A']))
print(stats.shapiro(df['group_B']))
print(stats.shapiro(df['group_C']))
print(stats.shapiro(df['group_D']))

# 2. Levene 검정 (등분산성)
print(stats.levene(df['group_A'], df['group_B'], df['group_C'], df['group_D']))

# 3. 일원 분석 분산
print(stats.f_oneway(df['group_A'], df['group_B'], df['group_C'], df['group_D']))
```

```python
# 데이터 재구조화
df_melt = df.melt(var_name="groups", value_name="scores")

# 일원분산분석을 위한 모델 학습
from statsmodels.formula.api import ols
model = ols('scores ~ groups', df_melt).fit()

# 4~9. ANOVA 테이블
from statsmodels.stats.anova import anova_lm
print(anova_lm(model))
```

### Section8.

```python
# 데이터
import pandas as pd
df = pd.DataFrame({
    '온도': [16, 20, 24, 28, 32, 18, 22, 26, 30, 34,
           23, 28, 25, 32, 21, 33, 24, 20, 24, 31],
    '판매량': [180, 210, 230, 280, 320, 370, 250, 240, 330, 380,
            245, 293, 237, 297, 234, 330, 243, 172, 231, 316]
})
print(df.head(3))
```

```python
from statsmodels.formula.api import ols

# 1. 상관계수
print("1. 상관계수:", round(df['온도'].corr(df['판매량']), 2))

# 회귀 모델
model = ols('판매량 ~ 온도', data=df).fit()

# 2. 결정계수
print("2. 결정계수(R-squared):", round(model.rsquared, 2))

# 3. 회귀계수(기울기)
print("3. 회귀계수:", round(model.params['온도'], 4))

# 4. 절편
print("4. 절편:", round(model.params['Intercept'], 4))

# 5. p-value 검정
print("5. pvalue:", round(model.pvalues['온도'], 4))

# 6. 온도가 25도일 때, 예측 판매량
new_data = pd.DataFrame({"온도": [25]})
result = model.predict(new_data)
print("6. 25도:", int(result[0]))

# 7. 잔차 제곱합
df['잔차'] = df['판매량'] - model.predict(df)
print("7. 잔차 제곱합:", round(sum(df['잔차']**2), 2))

# 8. MSE(Mean Squared Error)
MSE = (df['잔차'] ** 2).mean()
print('8. MSE:', round(MSE, 4))

# 9. 기울기(회귀 계수)에 대한 95% 신뢰구간
print("9. 신뢰구간:\n", model.conf_int().loc['온도'])

# 10. 온도가 35도 일 때, 95% 신뢰구간과 예측구간
new_data = pd.DataFrame({"온도": [35]})
pred = model.get_prediction(new_data)
result = pred.summary_frame(alpha=0.05)
print("10. 예측값의 신뢰구간과 예측구간:\n", result)

# 선형회귀 summary 결과로 2~5문제의 답을 찾을 수 있다.
print(model.summary())
```

### Section9.

```python
# 데이터
import pandas as pd
df = pd.DataFrame({
    '할인율': [28, 24, 13, 0, 27, 30, 10, 16, 6, 5, 7, 11, 11, 30, 25,
            4, 7, 24, 19, 21, 6, 10, 26, 13, 15, 6, 12, 6, 20, 2],
    '온도': [15, 34, 15, 22, 29, 30, 14, 17, 28, 29, 19, 19, 34, 10,
           29, 28, 12, 25, 32, 28, 22, 16, 30, 11, 16, 18, 16, 33, 12, 22],
    '광고비': [342, 666, 224, 764, 148, 499, 711, 596, 797, 484, 986, 347, 146, 362, 642,
            591, 846, 260, 560, 941, 469, 309, 730, 305, 892, 147, 887, 526, 525, 884],
    '주문량': [635, 958, 525, 25, 607, 872, 858, 732, 1082, 863, 904, 686, 699, 615, 893,
            830, 856, 679, 918, 951, 789, 583, 988, 631, 866, 549, 910, 946, 647, 943]
})
print(df.head(3))
```

```python
# 다중 선형 회귀 모델 적합
from statsmodels.formula.api import ols
model = ols('주문량 ~ 할인율 + 온도 + 광고비', data=df).fit()

# 1. 상관계수
print("1. 상관계수:", round(df['할인율'].corr(df['온도']), 2))

# 2. 결정계수
print("2. 결정계수(R-squared):", round(model.rsquared, 2))

# 3. 회귀계수(기울기)
print("3. 회귀계수:", round(model.params, 4))

# 4. 절편
print("4. 절편:", round(model.params['Intercept'], 4))

# 5. 회귀계수 검정
print("5. pvalue:", round(model.pvalues['온도'], 4))

# 6. 예측 판매량
new_data = pd.DataFrame({"할인율": [10], "온도": [20], "광고비": [500]})
result = model.predict(new_data)
print("6. 새로운 데이터:", int(result[0]))

# 7. 잔차 제곱합
df['잔차'] = df['주문량'] - model.predict(df)
print("7. 잔차 제곱합:", round(sum(df['잔차']**2), 2))

# 8. MSE(Mean Squared Error)
MSE = (df['잔차'] ** 2).mean()
print('8. MSE:', round(MSE, 4))

# 9. 각 변수에 대한 90% 신뢰구간
print("9. 신뢰구간:\n", model.conf_int(alpha=0.1))

# 10. 새로운 데이터의 예측값의 90% 신뢰구간과 예측구간
new_data = pd.DataFrame({"할인율": [15], "온도": [25], "광고비": [300]})
pred = model.get_prediction(new_data)
result = pred.summary_frame(alpha=0.1)
print("10. 예측값의 신뢰구간과 예측구간:\n", result)

# 11. 광고비는 배달 주문량에 영향을 주는지 가설 검정
cond = model.pvalues['광고비'] < 0.05
if cond:
    result = "기각"
else:
    result = "채택"
print("11. 귀무가설", result)

# 12. 광고비를 올리면 배달 주문량이 증가한다는 가설 검정 (단측검정)
cond1 = model.params['광고비'] > 0
cond2 = model.pvalues['광고비'] / 2 < 0.05
if cond1 and cond2:
    result = "기각"
else:
    result = "채택"
print("12. 귀무가설", result)

# 선형 회귀 모델의 요약 결과
print(model.summary())
```
